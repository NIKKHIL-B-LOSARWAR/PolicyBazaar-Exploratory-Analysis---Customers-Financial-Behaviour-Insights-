# **Problem Statement**

## **1. Company's Overview & Business Context**

Founded in 2008, **PolicyBazaar** emerged to address the inherent complexities and lack of transparency within the Indian insurance and financial services market. Recognizing the challenges faced by consumers in navigating a confusing landscape of financial products and providers, the company initially focused on simplifying insurance comparisons and empowering individuals with the information necessary to make informed decisions. Over time, PolicyBazaar's vision expanded, evolving into a comprehensive platform that encompasses a broader spectrum of financial products, including loans, credit cards, and investments. PolicyBazaar has established itself as a market leader in the online insurance and financial services sector in India. The company facilitates the entire customer journey, from product comparison and expert assistance to purchase, documentation, and transaction tracking, all within a single platform.

## **2. Project's Objective**

In the competitive landscape of financial services, understanding customer behavior is crucial for enhancing customer experience, improving risk management, and optimizing financial product offerings. This project aims to perform an exploratory data analysis (EDA) on the PolicyBazaar dataset to uncover patterns and insights into customers' financial behavior.

## **Focused areas for Analysis -**

### **1. Customer Demographics and Financial Behavior:**

   In this area of analysis, we assess PolicyBazaar's customer base by segmenting them into four distinct groups to enable highly relevant and optimized financial product recommendations. The segmentation is driven by key financial metrics, primarily monthly in-hand salary, annual income, and monthly bank balance. This approach provides deeper insights into customers' financial stability and purchasing power, ensuring tailored financial solutions that align with their specific needs and capabilities.


### **2. Credit Score and Risk Management:**


  This analysis evaluates the credit health of PolicyBazaar’s customer base by identifying high-risk segments based on their credit utilization and repayment behaviors. Customers with excessive credit utilization, late payments, and frequent credit inquiries are assessed to determine their potential credit default risk. Additionally, insights into age groups and professions prone to high credit usage help refine risk assessment strategies. By categorizing customers based on payback patterns and identifying those with credit scores critically at risk, this analysis enables targeted financial guidance and responsible lending practices, ultimately improving PolicyBazaar’s ability to offer well-suited financial products while mitigating default risks.


### **3. Loan and Credit Analysis:**

This analysis focuses on assessing debt and loan-related risks within PolicyBazaar’s customer base to identify high-risk borrowers, evaluate DTI (Debt-to-Income) segmentation, and determine ideal customer segments for financial product recommendations. By analyzing loan default risk factors, delayed payments, and small payback behaviors, we pinpoint customers prone to repayment challenges. Additionally, the study identifies high-revenue customers and assesses their contribution to the total debt capital issued. Insights into top professions and age groups with significant interest payments help refine revenue-based segmentation. These findings enhance loan structuring, risk mitigation, and personalized financial recommendations, optimizing both customer satisfaction and business profitability.


### **4. Investment and Savings Patterns:**

This analysis delves into investment behavior patterns within PolicyBazaar’s customer base to identify high-net-worth individuals, assess investment trends, and determine key demographic segments actively investing. By segmenting customers based on net worth and monthly investment amounts, we highlight the top 25% of investors and their contribution to total invested capital. Additionally, evaluating investment trends across months, professions, and age groups helps pinpoint periods of high engagement and customer profiles with strong financial planning habits. Correlation analysis further uncovers key financial factors influencing investment behavior, enabling targeted recommendations for wealth management solutions and optimizing PolicyBazaar’s investment-related offerings.







In summary,

By leveraging this dataset, the project aims to provide actionable insights into customer segments, identify risk factors related to credit scores, and enhance financial decision-making processes. This analysis will support PolicyBazaar in tailoring financial products, improving customer satisfaction, and mitigating risks associated with lending and credit.
